## Example usage of `DataSet` class

```python
import numpy as np
import katcali
import pylab as plt
print(katcali.__file__)
```

```python
# Load dataset
ds = katcali.dataset.DataSet(dbname="../data/meerkat2019.json")
```

```python
# Available files
ds.files.keys()
```

```python
fname = list(ds.files.keys())[0]
print("Filename:", fname)
```

```python
# Get metadata for this file
meta0 = ds.get_metadata(fname)
```

```python
ds.get_bad_ants(fname)
```

```python
vis, flags = ds.load_data(fname, recv='m006', pol='h')
```

```python
print(vis.shape)
plt.matshow(vis, cmap='viridis', aspect='auto')
plt.colorbar()
plt.gcf().set_size_inches((14., 8.))
```
